```python
from statsmodels.tsa.stattools import adfuller#ADF Test


# Define the return series and their names
return_cols = [
    df['R_GOLD'],
    df['R_PLATINUM'],
    df['R_NATURALGAS'],
    df['R_CRUDEOIL'],
    df['R_COFFEE']
]
return_names = ['Gold Return', 'Platinum Return', 'Natural Gas Return', 'Crude Oil Return', 'Coffee Return']

# Dictionary to store results
adf_results = {}

for i, col_series in enumerate(return_cols):
    # Drop NaN values
    cleaned_series = col_series.dropna()
    if len(cleaned_series) > 0:
        adf_test = adfuller(cleaned_series)
        adf_stat = adf_test[0]
        p_value = adf_test[1]
        critical_value = adf_test[4]['5%']
        stationary = "Yes" if p_value < 0.05 else "No"
        adf_results[return_names[i]] = {
            'ADF Statistic': adf_stat,
            'p-value': p_value,
            '5% Critical Value': critical_value,
            'Stationary (5%)': stationary
        }
    else:
        adf_results[return_names[i]] = {
            'ADF Statistic': None,
            'p-value': None,
            '5% Critical Value': None,
            'Stationary (5%)': 'Series is empty'
        }

# Convert dictionary to DataFrame and display
adf_df = pd.DataFrame(adf_results).T
display(adf_df)
```

```python
import numpy as np

# Compute log returns
for col in price_col:
    df[f'{col}_log_return'] = np.log(df[col] / df[col].shift(1))

# Drop NaN rows created by shifting
df_returns = df.dropna()
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox#Ljungbox Test
import pandas as pd

# Define the return series and their names
return_cols = [
    df['R_GOLD'],
    df['R_PLATINUM'],
    df['R_NATURALGAS'],
    df['R_CRUDEOIL'],
    df['R_COFFEE']
]
return_names = ['Gold Return', 'Platinum Return', 'Natural Gas Return', 'Crude Oil Return', 'Coffee Return']

# Dictionary to store results
ljung_results = {}

for i, col_series in enumerate(return_cols):
    # Drop NaN values
    cleaned_series = col_series.dropna()
    if len(cleaned_series) > 0:
        lb_test = acorr_ljungbox(cleaned_series, lags=[10], return_df=True)  # 10 lags example
        stat = lb_test['lb_stat'].values[0]
        p_value = lb_test['lb_pvalue'].values[0]
        autocorr = "Yes" if p_value < 0.05 else "No"
        ljung_results[return_names[i]] = {
            'Ljung-Box Statistic (lag 10)': stat,
            'p-value': p_value,
            'Autocorrelation at 5%?': autocorr
        }
    else:
        ljung_results[return_names[i]] = {
            'Ljung-Box Statistic (lag 10)': None,
            'p-value': None,
            'Autocorrelation at 5%?': 'Series is empty'
        }

# Convert dictionary to DataFrame and display
ljung_df = pd.DataFrame(ljung_results).T
display(ljung_df)
```

```python
import pandas as pd
import scipy.stats as stats

# Assuming df already has your return columns and no null values
return_cols = ['R_GOLD','R_PLATINUM','R_NATURALGAS','R_CRUDEOIL'	,'R_COFFEE']  # Replace with your actual return columns

# Prepare results list
summary_stats = []

for col in return_cols:
    series = df[col].dropna()
    mean = series.mean()
    std = series.std()
    skewness = series.skew()
    kurtosis = series.kurtosis()

    # One-sample t-test (testing against mean=0)
    t_stat, p_val = stats.ttest_1samp(series, 0)

    summary_stats.append({
        "Return Series": col,
        "Mean": mean,
        "Std Dev": std,
        "Skewness": skewness,
        "Kurtosis": kurtosis,
        "Mean t-Statistic": t_stat,
        "Mean p-Value": p_val
    })

# Create the summary table
summary_table = pd.DataFrame(summary_stats)
print(summary_table)
```

```python
from scipy.stats import jarque_bera


# Define the return series and their names
return_cols = [
    df['R_GOLD'],
    df['R_PLATINUM'],
    df['R_NATURALGAS'],
    df['R_CRUDEOIL'],
    df['R_COFFEE']
]
return_names = ['Gold Return', 'Platinum Return', 'Natural Gas Return', 'Crude Oil Return', 'Coffee Return']

# Dictionary to store results
jarque_bera_results = {}

# Loop through each return series
for i, col_series in enumerate(return_cols):
    # Drop NaN values before performing the test
    cleaned_series = col_series.dropna()
    if len(cleaned_series) > 0:
        jb_stat, jb_p_value = jarque_bera(cleaned_series)
        jarque_bera_results[return_names[i]] = {
            'Jarque-Bera Statistic': jb_stat,
            'p-value': jb_p_value
        }
    else:
        jarque_bera_results[return_names[i]] = {
            'Jarque-Bera Statistic': None,
            'p-value': None,
            'Note': 'Series is empty after dropping NaN values'
        }

# Convert the results to a DataFrame and display
jarque_bera_df = pd.DataFrame(jarque_bera_results).T
display(jarque_bera_df)
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy.stats import jarque_bera

# summary format of:
# 1️⃣ ADF Test for Price Series
# -----------------------------
price_cols = ['GOLD ', 'PLATINUM', 'NATURAL GAS', 'CRUDE OIL', 'COFFEE']
adf_results = []

for col in price_cols:
    series = df[col].dropna()
    adf_test = adfuller(series)
    p_value = adf_test[1]
    stationary = "Yes" if p_value < 0.05 else "No"

    adf_results.append({
        "Commodity": col,
        "ADF Statistic": adf_test[0],
        "p-value": p_value,
        "5% Critical Value": adf_test[4]['5%'],
        "Stationary at 5%?": stationary
    })

adf_table = pd.DataFrame(adf_results)

# -----------------------------
# 2️⃣ Jarque-Bera Test for Return Series
# -----------------------------
return_cols = ['R_GOLD', 'R_PLATINUM', 'R_NATURALGAS', 'R_CRUDEOIL', 'R_COFFEE']
jb_results = []

for col in return_cols:
    series = df[col].dropna()
    jb_stat, jb_p = jarque_bera(series)
    normal = "Yes" if jb_p > 0.05 else "No"

    jb_results.append({
        "Commodity": col,
        "JB Statistic": jb_stat,
        "p-value": jb_p,
        "Normally Distributed at 5%?": normal
    })

jb_table = pd.DataFrame(jb_results)

# -----------------------------
# 3️⃣ Ljung-Box Test for Return Series
# -----------------------------
lb_results = []

for col in return_cols:
    series = df[col].dropna()
    lb_test = acorr_ljungbox(series, lags=[10], return_df=True)
    lb_stat = lb_test['lb_stat'].values[0]
    lb_p = lb_test['lb_pvalue'].values[0]
    independent = "Yes" if lb_p > 0.05 else "No"

    lb_results.append({
        "Commodity": col,
        "LB Statistic": lb_stat,
        "p-value": lb_p,
        "Independent at 5%?": independent
    })

lb_table = pd.DataFrame(lb_results)

# -----------------------------
# Display all tables
# -----------------------------
display(adf_table)
display(jb_table)
display(lb_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the y-axis column
y_col = 'R_GOLD'

# Define the x-axis columns (all other return columns)
x_cols = ['R_PLATINUM','R_NATURALGAS','R_CRUDE OIL','R_COFFEE']

# Create a scatter plot for R_GOLD against each of the other return series
plt.figure(figsize=(15, 10))
for i, col in enumerate(x_cols):
    plt.subplot(2, 2, i + 1)  # Arrange plots in a 2x2 grid (adjust as needed)
    sns.scatterplot(x=df[col], y=df[y_col])
    plt.title(f'Scatter Plot: {y_col} vs {col}')
    plt.xlabel(col)
    plt.ylabel(y_col)
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Define dependent and independent variables
# Select relevant columns and drop rows with any NaN values in these columns
regression_cols = ['R_GOLD', 'R_PLATINUM', 'R_NATURALGAS', 'R_CRUDE OIL', 'R_COFFEE']
df_cleaned = df[regression_cols].dropna()

y = df_cleaned['R_GOLD']
x = sm.add_constant(df_cleaned[['R_PLATINUM','R_NATURALGAS','R_CRUDE OIL','R_COFFEE']])
```

```python
# Fit the OLS model
model = sm.OLS(y, x).fit()
```

```python
# Print the regression results
print(model.summary())
```

```python

# Diagnostic Tests

# 1. Residual Plot
plt.figure(figsize=(8, 6))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
# 2. Histogram of Residuals
plt.figure(figsize=(8, 6))
plt.hist(model.resid, bins=20)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
# 3. Q-Q Plot
plt.figure(figsize=(8, 6))
sm.qqplot(model.resid, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
# 6. Durbin-Watson Test for Autocorrelation
dw_test = sm.stats.durbin_watson(model.resid)
print("\nDurbin-Watson Test:")
print(f"Durbin-Watson statistic: {dw_test:.3f}")

from statsmodels.stats.diagnostic import het_breuschpagan
```

```python
# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)

# Extract the test statistics
bp_test_statistic, bp_p_value, _, _ = bp_test

print(f"Breusch-Pagan test statistic: {bp_test_statistic}")
print(f"p-value: {bp_p_value}")
```
